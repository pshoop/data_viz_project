```python
# Import all the things
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster

import json

# Make plots inline
%matplotlib inline
```

```python
# Read in the data that conveniently has lat and long data
df_beers = pd.read_csv('./data/df_beers.csv')
df_beers.head()
```

```python
# Drop any and all NaNs because I am being to lazy to fix them with accurate data
df_beers.dropna(inplace=True)

# Create a new dataframe with just the geolocation data
df_locs = df_beers[['brewery_name', 'latitude', 'longitude']].copy()
```

```python
# lotsa dupes here (because each brewery has a different entry for each beer style)
# Get rid of the dupes (but keep 1!)
df_locs.drop_duplicates(subset='brewery_name', keep='first', inplace=True)
```

```python
heat_data = [[row['latitude'],row['longitude']] for index, row in df_locs.iterrows()]



# Create a Map instance
m = folium.Map(location=[10, -40], zoom_start=2, control_scale=True)

# Add heatmap to map instance
# Available parameters: HeatMap(data, name=None, min_opacity=0.5, max_zoom=18, max_val=1.0, radius=25, blur=15, gradient=None, overlay=True, control=True, show=True)
HeatMap(heat_data).add_to(m)

# Alternative syntax:
#m.add_child(HeatMap(points_array, radius=15))

# Show map
m.save('./plots/heatmap.html')
m
```

```python
# Import a library I forgot
from folium.plugins import FastMarkerCluster

# Create a center point for the map
rome_lat, rome_lng = 41.9028, 12.4964

# Create a map shell centered on Rome... why Rome? Cause it was the example I used....
# and I'm too lazy to look up lat/long for Pittsburgh.
my_map = folium.Map(location=[rome_lat, rome_lng], zoom_start=2)

# Fill my map shell with markers for each brewery.
my_map.add_child(FastMarkerCluster(df_locs[['latitude', 'longitude']].values.tolist()))

# Save the file
my_map.save("./plots/marker_map.html")
my_map
```
